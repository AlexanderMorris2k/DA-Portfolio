## Task 2

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('food_claims_2212.csv')

location_counts = data['location'].value_counts()

location_counts.plot(kind='bar')

plt.title('Number of Claims by Location')
plt.xlabel('Location')
plt.ylabel('Number of Claims')

plt.show()

print("Number of Entries by Location:")
for location, count in location_counts.items():
    print(f"{location}: {count}")

print('The Recife location has the most claims')

print('The number of claims for each location is not balanced')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('food_claims_2212.csv')

time_to_close_data = data['time_to_close']
location_data = data['location']

plt.scatter(time_to_close_data, location_data)

plt.title('Relationship between Time to Close and Location')
plt.xlabel('Time to Close')
plt.ylabel('Location')

plt.show()

# The straight line shows that there is a strong linear relationship between the time to close and the claim location
```
